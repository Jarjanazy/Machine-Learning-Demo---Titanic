```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")
```

## Perform data analysis

```python
count_of_females_survived = len(df [ (df['Survived'] == 1) & (df['Sex'] == 'female') ])
count_of_males_survived = len(df [ (df['Survived'] == 1) & (df['Sex'] == 'male') ])

count_of_females_dead = len(df [ (df['Survived'] == 0) & (df['Sex'] == 'female') ])
count_of_males_dead = len(df [ (df['Survived'] == 0) & (df['Sex'] == 'male') ])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].set_title('Surivaval by sex')
ax[0].pie([count_of_females_survived, count_of_males_survived], labels=['Females', 'Male'], autopct='%1.1f%%')

ax[1].set_title('Death by sex')
ax[1].pie([count_of_females_dead, count_of_males_dead], labels=['Females', 'Male'], autopct='%1.1f%%')

plt.show()
```

```python
age_survival_histogram = df[ df[ 'Survived'] == 1 ] ['Age']
age_death_histogram = df[ df[ 'Survived'] == 0 ] ['Age']
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharey=True)


ax[0].set_title('Survival by age')
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Count of Survivals')
ax[0].hist(age_survival_histogram, bins = 20,  edgecolor='black')

ax[1].set_title('Death by age')
ax[1].set_xlabel('Age')
ax[1].set_ylabel('Count of Survivals')
ax[1].hist(age_death_histogram, bins = 20,  edgecolor='black')

plt.show()
```

```python
pclass_survival = df[ df['Survived'] == 1]['Pclass']
pclass_death= df [ df['Survived'] == 0 ]['Pclass']


pclass_survival_values_count_seris = pclass_survival.value_counts().sort_index()
pclass_death_values_count_seris = pclass_death.value_counts().sort_index()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1)


ax[0].set_title('Survival by Pclass')
ax[0].pie(pclass_survival_values_count_seris.values, labels = pclass_survival_values_count_seris.index, autopct='%1.1f%%')

ax[1].set_title('Death by Pclass')
ax[1].pie(pclass_death_values_count_seris.values, labels = pclass_death_values_count_seris.index, autopct='%1.1f%%')

plt.show()
```

```python
embarked_survival = df[ df['Survived'] == 1]['Embarked']
embarked_death= df [ df['Survived'] == 0 ]['Embarked']


embarked_survival_values_count_seris = embarked_survival.value_counts().sort_index()
embarked_death_values_count_seris = embarked_death.value_counts().sort_index()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1)


ax[0].set_title('Survival by Embarked')
ax[0].pie(embarked_survival_values_count_seris.values, labels = embarked_survival_values_count_seris.index, autopct='%1.1f%%')

ax[1].set_title('Death by Embarked')
ax[1].pie(embarked_death_values_count_seris.values, labels = embarked_death_values_count_seris.index, autopct='%1.1f%%')

plt.show()
```

```python
fare_survival = df[ df['Survived'] == 1]['Fare']
fare_death= df [ df['Survived'] == 0 ]['Fare']
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharey=True)


ax[0].set_title('Survival by Fare')
ax[0].set_xlabel('Fare')
ax[0].set_ylabel('Count of Survivals')
ax[0].hist(fare_survival, bins = 15,  edgecolor='black')

ax[1].set_title('Death by Fare')
ax[1].set_xlabel('Fare')
ax[1].set_ylabel('Count of Death')
ax[1].hist(fare_death, bins = 15,  edgecolor='black')

plt.show()
```
